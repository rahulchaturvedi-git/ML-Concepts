```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define the data points
x = np.array([1,2,-1,-2]) #feature values
y = np.array([1,1,0,0]) # corresponding labels
```

```python
#sigmoid function
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
#cross entropy loss function
def cross_entropy_loss(x, y, theta, theta0):
  z = theta*x + theta0
  p = sigmoid(z)
  loss = -np.mean(y*np.log(p  + 1e-15) + (1-y)*np.log(1-p + 1e-15))
  return loss
```

```python
#values of theta and theta0 to evaluate the loss function
theta_values = [1,-1,2,-2] #different thetas
theta0_values = [0,1,-1,2] #corresponding theta0s
```

```python
#plot setup
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Sigmoid fucntion and cross entropy loss fucntion for different parameters', fontsize=16)

for ax, theta, theta0 in zip(axes.flatten(), theta_values, theta0_values):
  #genearte a range of x values for plotting the sigmoid
  x_range = np.linspace(-3,3,100)
  z = theta*x_range + theta0
  p = sigmoid(z)

  #plot sigmoid function
  ax.plot(x_range, p, label=f"Sigmoid: θ={theta}, θ0={theta0}")

  #highlight the data point
  ax.scatter(x, np.zeros_like(x), c = ['blue' if label == 1 else 'red' for label in y],s=50)
  ax.vlines(x,0,sigmoid(theta * x + theta0), color = 'green', linestyles = 'dotted')

  #calculate and display the loss
  loss = loss = cross_entropy_loss(x, y, theta, theta0)
  ax.title.set_text(f"Theta = {theta}, Theta0 = {theta0} | Cross Entropy Loss: {loss:.4f}")

  ax.set_xlim([-3,3])
  ax.set_ylim([-0.1,1.1])
  ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
