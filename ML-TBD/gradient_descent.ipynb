```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(grad_f, start_x, learning_rate=0.1, tolerance=1e-6):
    x = start_x
    iteration = 0

    history = [x]

    while True:
        gradient = grad_f(x)

        if abs(gradient) < tolerance:
            break

        x = x - learning_rate * gradient
        history.append(x)

        if iteration % 10 == 0:
            print(f"Iteration {iteration}: x = {x}, gradient = {gradient}")

        iteration += 1

    return x, history


# f(x) = x²
def f(x):
    return x**2


# f'(x) = 2x
def grad(x):
    return 2 * x


minimum, history = gradient_descent(
    grad,
    start_x=10,
    learning_rate=0.1
)

print("\nApproximate minimum at:", minimum)

# Plot the function
x_curve = np.linspace(-11, 11, 500)
y_curve = f(x_curve)

plt.figure(figsize=(8, 5))
plt.plot(x_curve, y_curve, label="f(x) = x²")

# Plot gradient descent path
history = np.array(history)
plt.scatter(
    history,
    f(history),
    s=50,
    label="Gradient Descent Steps"
)

# Connect the points
plt.plot(
    history,
    f(history),
    linestyle="--"
)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent on f(x) = x²")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.figure(figsize=(6, 6))
contours = plt.contour(X, Y, Z, levels=20)
plt.clabel(contours, inline=True, fontsize=8)

plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Contours of $f(x,y)=x^2+y^2$')
plt.axis('equal')
plt.grid(True)

plt.show()
```

```python
import numpy as np

def gradient_descent_2d(grad_f, start_x, start_y,
                        learning_rate=0.1,
                        tolerance=1e-6):

    x = start_x
    y = start_y
    iteration = 0

    history = [(x, y)]

    while True:

        grad_x, grad_y = grad_f(x, y)

        # magnitude of gradient vector
        grad_norm = np.sqrt(grad_x**2 + grad_y**2)

        if grad_norm < tolerance:
            break

        x = x - learning_rate * grad_x
        y = y - learning_rate * grad_y

        history.append((x, y))

        if iteration % 10 == 0:
            print(
                f"Iteration {iteration}: "
                f"x={x:.6f}, y={y:.6f}, "
                f"|grad|={grad_norm:.6f}"
            )

        iteration += 1

    return x, y, history


# Gradient of f(x,y)=x²+y²
def grad(x, y):
    return 2*x, 2*y


x_min, y_min, history = gradient_descent_2d(
    grad,
    start_x=10,
    start_y=8
)

print("\nApproximate minimum:")
print(f"x = {x_min}")
print(f"y = {y_min}")
```
