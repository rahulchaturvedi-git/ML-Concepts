```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(grad_f, start_x, learning_rate=0.1, tolerance=1e-6):
    x = start_x
    iteration = 0

    history = [x]

    while True:
        gradient = grad_f(x)

        if abs(gradient) < tolerance:
            break

        x = x - learning_rate * gradient
        history.append(x)

        if iteration % 10 == 0:
            print(f"Iteration {iteration}: x = {x}, gradient = {gradient}")

        iteration += 1

    return x, history


# f(x) = x²
def f(x):
    return x**2


# f'(x) = 2x
def grad(x):
    return 2 * x


minimum, history = gradient_descent(
    grad,
    start_x=10,
    learning_rate=0.1
)

print("\nApproximate minimum at:", minimum)

# Plot the function
x_curve = np.linspace(-11, 11, 500)
y_curve = f(x_curve)

plt.figure(figsize=(8, 5))
plt.plot(x_curve, y_curve, label="f(x) = x²")

# Plot gradient descent path
history = np.array(history)
plt.scatter(
    history,
    f(history),
    s=50,
    label="Gradient Descent Steps"
)

# Connect the points
plt.plot(
    history,
    f(history),
    linestyle="--"
)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent on f(x) = x²")
plt.legend()
plt.grid(True)

plt.show()
```
